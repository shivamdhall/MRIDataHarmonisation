```python
def plot_losses(training_losses, validation_losses,nine, twelve, title):
    # This function plots the training losses and validation losses after each epoch

    fig = plt.figure()
    plt.plot(range(1,len(training_losses)+1), training_losses, 'r-', label="5x5x5")
    plt.plot(range(1,len(validation_losses)+1), validation_losses, 'b-', label="7x7x7")
    plt.plot(range(1,len(nine)+1), nine, 'g-', label="9x9x9")
    plt.plot(range(1,len(twelve)+1), twelve, 'y-', label="11x11x11")
    plt.legend()
    plt.xlabel('epoch')
    # Make the y-axis label, ticks and tick labels match the line color.
    plt.ylabel('Loss - (MSE)')
    plt.title(title)
    plt.show()
```
