```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def plot_losses(training_losses, validation_losses, title):
    # This function plots the training losses and validation losses after each epoch

    fig = plt.figure()
    plt.plot(range(1,len(training_losses)+1), training_losses, 'r-', label="Training error")
    plt.plot(range(1,len(validation_losses)+1), validation_losses, 'b-', label="Validation error")
    plt.legend()
    plt.xlabel('epoch')
    # Make the y-axis label, ticks and tick labels match the line color.
    plt.ylabel('Loss - (MSE)')
    plt.title(title)
    plt.show()
```

```python
training_losses = [11100, 4800, 4400, 3984, 2300, 2500, 2200, 1950, 1700, 1680, 1720, 1700, 1340, 1400, 1528, 1495, 1450, 1420, 1400, 1445, 1400, 1383, 1340, 1330, 1335]
```

```python
plot_losses(training_losses, training_losses, "test")
```
